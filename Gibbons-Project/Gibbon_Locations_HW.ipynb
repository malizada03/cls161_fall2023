```python
#parsers 
#loading out NER parser 

#name entity recognition, AI model, identify things in text
nlp_ner= stanza.Pipeline(lang='en', processors='tokenize, ner',tokenize_pretokenized=True)
```

```python
#if you want to pretokenize and split your text by sentence 
tokenized_chapter= mychapter.split('. ')
#replace my chapter with your actual variable 

#print(tokenized_chapter)
```

```python
#creating a stanza document 
#can take 3-6 mins depending on ur computer setup 
#if you have pre_tokenized the text 
Docs = [stanza.Document([], text=d) for d in tokenized_chapter]
Out_Docs = nlp_ner(Docs)  
 
```

```python
#accessing named entities for pre tokenized out_docs 
for doc in Out_Docs:
    print(doc.ents) 
```

```python
#how to add values to the dictionary, create conditions so that only location tokens go into dictionary 
count = 0
NER_Dict= {}    #token.text will show the text of each word, #token.ner will show NER, will == 'O' if token is not a named entity
for doc in Out_Docs:
    for sent in doc.sentences:
        for token in sent.tokens:
            if "LOC" in token.ner:#my code
                count = count + 1
                print(f'{token.text},{token.ner}')
            
#create a dictionary (NER_Dict) where the key is the token text and the value is the named entity tag 
#only want locations from Gibbons that will later be mapped
```

```python
#worked with Micah to develop this code on 12/21 zoom meeting!
#create a dictionary (NER_Dict) where the key is the token text and the value is the named entity tag
NER_Dict2= {}    #token.text will show the text of each word, #token.ner will show NER, will == 'O' if token is not a named entity
for doc in Out_Docs:
    for sent in doc.sentences:
        for token in sent.tokens:
            # print(f'{token.text},{token.ner}')
            if token.ner == 'O':  # note that this is the letter and not a zero
              continue  # can you guess what "continue" does?
            else:
              NER_Dict2[token.text] = token.ner
```

```python
ner_frame= pd.DataFrame.from_dict(NER_Dict2, orient= 'index')  #creates a pandas dataframe from dictionary 
ner_frame.head()
```
